```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

Predict pizza Category on the basis of unit price and Pizza Size

```python
df = pd.read_csv('pizza_sales.csv')
```

```python
df.columns
```

```python
df['pizza_size'].unique()
```

```python
pizza_size_id = {
    'M':1,
    'L':2,
    'S':0,
    'XL':3,
    'XXL':4
}
df['pizza_size_id'] = df['pizza_size'].map(pizza_size_id)
```

```python
df
```

```python
le = LabelEncoder()
```

```python
df['pizza_category_id']= le.fit_transform(df['pizza_category'])
```

```python
features = ['unit_price','pizza_size_id']
target = 'pizza_category_id'
```

```python
X = df[features]
Y = df[target]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(
    X,Y,test_size = 0.2,random_state = 42
)
```

```python
scaler = StandardScaler() #Scaling should only be done on features
```

```python
#xi-mean/sd
X_train_scale = scaler.fit_transform(X_train)
X_train_scale
```

```python
X_test_scale = scaler.transform(X_test)
X_test_scale
```

```python
error_rate =[]
k_range = range(1,21)
```

```python
for i in k_range:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scale,Y_train)
    Y_pred = knn.predict(X_test_scale)
    error_rate.append(np.mean(Y_pred != Y_test)) #error rate = sum of wrong values/total no of predicted values
```

```python
plt.plot(k_range,error_rate,marker = 'o')
```

```python
k = 17
```

```python
model = KNeighborsClassifier(n_neighbors =k )                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             
```

```python
model.fit(X_train_scale,Y_train)
```

```python
Y_pred =model.predict(X_test_scale)
```

```python
cr = classification_report(Y_pred,Y_test)
print(cr)
```

```python
new_data = {
    'unit_price':[13.5,20],
    'pizza_size_id':[0,4]
}
X_new = pd.DataFrame(new_data)
```

```python
X_new_scale = scaler.transform(X_new)
X_new_scale
```

```python
new_pred = model.predict(X_new_scale)
```

```python
new_pred
```

```python
category_value = le.inverse_transform(new_pred)
category_value
```

```python
X_new['Predicted Category'] = category_value
X_new
```

```python
data = pd.read_csv('HR_Dataset Refresh.csv')
```

```python
data.columns
```

```python
data['Department'].unique()
```

```python
department_id = {
      'Production       ':0,
    'IT/IS':1,
    'Software Engineering':2,
    'Admin Offices':3,
    'Sales':4,
    'Executive Office':5
}
data['Department_Id']= data['Department'].map(department_id)
```

```python
data['EmpSatisfaction'].unique()
```

```python
data['SpecialProjectsCount'].unique()
```

```python
data['PerformanceScore'].unique()
```

```python
data['Performance_score_id'] = le.fit_transform(data['PerformanceScore'])
data['Performance_score_id']
```

```python
features = ['Department_Id','Salary','EmpSatisfaction','SpecialProjectsCount']
target = 'Performance_score_id'
```

```python
X = data[features]
Y = data[target]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(
    X,Y,test_size = 0.2,random_state = 42
)
```

```python
scaler = StandardScaler()
```

```python
X_train_scale = scaler.fit_transform(X_train)
X_train_scale
```

```python
X_test_scale = scaler.transform(X_test)
X_test_scale
```

```python
error_rate =[]
k_range = (1,21)
```

```python
for i in k_range:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scale,Y_train)
    Y_pred = knn.predict(X_test_scale)
    error_rate.append(np.mean(Y_pred != Y_test)) #error rate = sum of wrong values/total no of predicted values
```

```python
plt.plot(k_range,error_rate,marker ='o')
```

```python
k = 21
```

```python
model = KNeighborsClassifier(n_neighbors= k)
```

```python
model.fit(X_train_scale,Y_train)
```

```python
Y_pred = model.predict(X_test_scale)
```

```python
cr = classification_report(Y_pred,Y_test)
print(cr)
```
