```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('Datas/pizza_sales.csv')
```

```python
df.head()
```

One Sample T-test

#Is the average total_price of all pizza sales different from $15

```python
'''Workflow-----------
define alpha value(risk limit)
define sample
define column name for sample comparison

testing -> ttest_1samp(column_name,popmean=)
it returns statistics and probablity

print results
compare pvalue and alpha value for correct hypothesis(assumption)
'''
```

```python
alpha = 0.05
sample = 15
total_sales = df['total_price']
```

```python
t_statistics, pvalue = stats.ttest_1samp(total_sales,popmean=sample)
```

```python
print(f"Ho: Average total price of all pizza is equal to ${sample}.")
print(f"Ha: Average total price of all pizza is not equal to ${sample}.")
print(f"Statistics:{t_statistics}")
print(f"p-value:{pvalue}")
```

```python
if pvalue < alpha:
    print(f"Decision:Failed to accept null Hypothesis.\nConclusion:Accept alternate Hypothesis:Ha: Average total price of all pizza is not equal to $15.")
else:
     print(f"Decision:Failed to reject null Hypothesis.\nConclusion:Accept null Hypothesis:Ha: Average total price of all pizza is not equal to $15.")
```

```python
#Is average salary of all employees is $5000?

data = pd.read_csv('HR_Dataset Refresh.csv')
```

```python
data.head()
```

```python
alp = 0.05
samp = 5000
avg_sal = data['Salary']
```

```python
T_statistics,p_value = stats.ttest_1samp(avg_sal,popmean=samp) 
```

```python
print(f"Ho: average salary of all employees is ${samp}")
print(f"Ha: average salary of all employees is not ${samp}")
```

```python
if p_value > alp:
    print(f"Null hypothesis ins true i.e Ho: average salary of all employees is ${samp}")
else:
    print(f"Alternate hypothesis is true i.e Ha: average salary of all employees is not ${samp}")
```
