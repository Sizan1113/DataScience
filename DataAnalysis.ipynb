Group By and Aggregate Function

```python
'''
Aggregate Function 
-------------------
Mean ->Calculates average record
Max->Extracts maximum value from column
Sum->Calculates total value for column
Min->Extracts minimum value from column
Count->Count total number of record
'''
```

Group By

```python
#Find total salary of employees by each department
total = df.groupby('DEPARTMENT_ID')['SALARY'].sum()
```

```python
total
```

```python
total = df.groupby('DEPARTMENT_ID')['SALARY'].sum().reset_index()
total
```

```python
total = df.groupby('DEPARTMENT_ID')['SALARY'].count().reset_index()
total
```

```python
#Find total salary ,average salary,max salary and min salary by job id
report = df.groupby('JOB_ID').agg(
    Total_Salary = ('SALARY','sum'),
    Average_Salary = ('SALARY','mean'),
    max_salary = ('SALARY','max'),
    min_salary = ('SALARY','min')
)
```

```python
report
```

```python
#Find total salary ,average salary,max salary and min salary by job id and get the index as numbers

report = df.groupby('JOB_ID').agg(
    Total_Salary = ('SALARY','sum'),
    Average_Salary = ('SALARY','mean'),
    max_salary = ('SALARY','max'),
    min_salary = ('SALARY','min')
).reset_index()
```

```python
report
```

```python
data = pd.read_csv("Datas/pizza_sales.csv")
data.head()
```

```python
data.columns
```

```python
#Find total number of pizza sold by pizza category
pizza_count = data.groupby('pizza_category')['order_id'].count().reset_index()
pizza_count
```

```python
pizzasize_count = data.groupby('pizza_size')['order_id'].count()
pizzasize_count
```

```python
#find total sales by each month
data['order_date']= pd.to_datetime(data['order_date'],format =  'mixed')
data['Month'] = data['order_date'].dt.month_name()
data.head()
```

```python
total_month = data.groupby('Month')['order_id'].count().reset_index()
total_month
```

```python
#Find total sales by each day
data['Day']=data['order_date'].dt.day_name()
data.head()
total_day = data.groupby('Day')['order_id'].count().reset_index()
total_day
```

```python
#Find total unit price,average unit price of pizza sales by pizza category
Details = data.groupby('pizza_category').agg(
    Total_unit_price = ('unit_price','sum'),
    Average_unit_price = ('unit_price','mean')
    
).reset_index()
```

```python
data = pd.read_csv("Datas/pizza_sales.csv")
data.head()
```

```python
data['order_time']= pd.to_datetime(data['order_time'], format='%H:%M:%S')
```

```python
data['order_date']= pd.to_datetime(data['order_date'],format =  'mixed')
data['Day']=data['order_date'].dt.day_name()
```

```python
weekday = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
data['Day']= pd.Categorical(data['Day'],categories = weekday)
```

```python
data.info()
```

```python
data['Hour']= data['order_time'].dt.hour
```

```python
data['Minute']= data['order_time'].dt.minute
```

```python
data['Second']= data['order_time'].dt.second
```

```python
data.head()
```

```python
Report = data.groupby(['Day','Hour'], observed=False).agg(
    Total_revenue_by_hour=('total_price','sum'),
    Average_revenue_by_hour=('total_price','mean')    
).reset_index()
```
