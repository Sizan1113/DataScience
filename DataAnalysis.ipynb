Group By and Aggregate Function

```python
'''
Aggregate Function 
-------------------
Mean ->Calculates average record
Max->Extracts maximum value from column
Sum->Calculates total value for column
Min->Extracts minimum value from column
Count->Count total number of record
'''
```

Group By

```python
#Find total salary of employees by each department
total = df.groupby('DEPARTMENT_ID')['SALARY'].sum()
```

```python
total
```

```python
total = df.groupby('DEPARTMENT_ID')['SALARY'].sum().reset_index()
total
```

```python
total = df.groupby('DEPARTMENT_ID')['SALARY'].count().reset_index()
total
```

```python
#Find total salary ,average salary,max salary and min salary by job id
report = df.groupby('JOB_ID').agg(
    Total_Salary = ('SALARY','sum'),
    Average_Salary = ('SALARY','mean'),
    max_salary = ('SALARY','max'),
    min_salary = ('SALARY','min')
)
```

```python
report
```

```python
#Find total salary ,average salary,max salary and min salary by job id and get the index as numbers

report = df.groupby('JOB_ID').agg(
    Total_Salary = ('SALARY','sum'),
    Average_Salary = ('SALARY','mean'),
    max_salary = ('SALARY','max'),
    min_salary = ('SALARY','min')
).reset_index()
```

```python
report
```
