```python
# Analysis Questions?
# --------------------------
# 1. Find total loan issued by each address. 
# 2. Find total loan issued by purpose and grade. 
# 3. Find average loan taken according to home_ownership. 
# 4. Find total issued loan by day and month. (multiple group by, sum)
# 5. Find total number of loan verified. (only show verified details)
#    -> groupby -> conditional filtering
# 6. Find average annual income by purpose.
```

```python
loan_address = data.groupby('address_state')['id'].count().sort_values(ascending=False).reset_index()
```

```python
loan_address
```

```python
filter_loanaddress = loan_address['id']>2000
```

```python
loan_address[filter_loanaddress]
```

```python
# 2. Find total loan issued by purpose and grade. 
loan_grade_purpose = data.groupby(['purpose','grade'])['id'].count().sort_values().reset_index()
```

```python
loan_grade_purpose
```

```python
#Find average loan taken according to home_ownership
loan_home = data.groupby('home_ownership')['id'].count().sort_values(ascending=False).reset_index()
```

```python
loan_home
```

```python
#Find total issued loan by day and month. (multiple group by, sum)
data['Issued_Month']=pd.to_datetime(data['issue_date']).dt.month_name()
data['Issued_Day']=pd.to_datetime(data['issue_date']).dt.day
```

```python
data.head()
```

```python
loan_day_month = data.groupby(['Issued_Month','Issued_Day'])['id'].count().reset_index()
```

```python
loan_day_month
```

```python
# Find total number of loan verified. (only show verified details)
#    -> groupby -> conditional filtering
Total_loan = data.groupby('verification_status')['id'].count().reset_index()
```

```python
Total_loan
```

```python
verified = (Total_loan['verification_status']=='Source Verified') & (Total_loan['verification_status']=='Verified')
Total_loan[verified]
```
