```python
import matplotlib.pyplot as plt
```

Line Plot

```python
fruits = ['Apple','Banana','orange','Coconut']
sales = [100,500,350,200]
revenue =[150,100,500,300]
 #X->independent variable
#y ->dependent variable
```

```python
plt.figure(figsize = (6,4))
plt.plot(fruits,sales,color='red' ,marker='o', label = 'Sales' )
plt.plot(fruits,revenue,color = 'green',marker = '*',label = 'Revenue',linestyle='--')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits and their sakes')
plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
plt.grid(True,linestyle='--', color = 'Blue', alpha=0.9)
plt.savefig('Fruits_sales.png')
plt.show()
```

Bar Chart

```python

plt.bar(fruits,sales,edgecolor= 'black' ,color=['orange','green','blue','red'])
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits and their sales')
#plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
plt.grid(True,linestyle='--', color = 'Blue', alpha=0.3)
plt.savefig('Fruits_sales.png')
plt.show()
```

```python
data = pd.read_csv("HR_Dataset Refresh.csv")
```

```python
data.head()
```

```python
plt.hist(data['Salary'], edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Salary frequency')
```

```python
import seaborn as sns
```

```python
mean_salary = data['Salary'].mean()
median_salary = data['Salary'].median()
```

```python
sns.histplot(data['Salary'],bins = 10,kde= True)
plt.axvline(mean_salary,color = 'red',linestyle='--',label='Mean')
plt.axvline(median_salary,color = 'yellow',label='Median')
```

```python
#Find total number of male and female and plot it in a bar graph
```

```python
count_gender = data['Gender'].value_counts().reset_index()
count_gender
```

```python
plt.bar(count_gender['Gender'],count_gender['count'])
plt.xlabel('Gender')
plt.ylabel('Total number of genders')
plt.title('Gender Counts')
for i,val in enumerate(count_gender['count']):
    plt.text(count_gender['Gender'][i],val+0.7 ,str(val))
```

```python
sns.barplot(x=count_gender['Gender'],y = count_gender['count'])
for i,val in enumerate(count_gender['count']):
    plt.text(count_gender['Gender'][i],val+0.7 ,str(val))
```

```python
#plot employees details-> performnces score wise salary according  to gender
ax=sns.barplot(x=data['PerformanceScore'],y=data['Salary'],palette='rocket_r',hue=data['Gender'],errorbar = None)
for container in ax.containers:
    ax.bar_label(container,fmt='%.0f')
```

```python
#Find salary according to Employee Satisfaction,performance Score,Department and plot it to different abr chart
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('HR_Dataset Refresh.csv')
data.head()
data.columns
```

```python
Total_Salary = data.groupby(['EmpSatisfaction','PerformanceScore','Department'])['Salary'].sum().reset_index()
```

```python
Total_Salary
```

```python
plt.bar(Total_Salary['Department'],Total_Salary['Salary'])
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=15)
plt.title('Department wise salary')
#plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
for i, val in enumerate(Total_Salary['Salary']):
    plt.text(Total_Salary['Department'][i], val, str(val))

plt.show()
```

```python
ax = sns.barplot(x=Total_Salary['Department'], y=Total_Salary['Salary'], errorbar=None)
plt.xticks(rotation=15)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

```python
plt.bar(Total_Salary['EmpSatisfaction'],Total_Salary['Salary'])
plt.xlabel('EmpSatisfaction')
plt.ylabel('Salary')
plt.xticks(rotation=15)
plt.title('EmpSatisfaction wise salary')
#plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
for i, val in enumerate(Total_Salary['Salary']):
    plt.text(Total_Salary['EmpSatisfaction'][i], val, str(val))

plt.show()
```

```python
plt.bar(Total_Salary['Department'],Total_Salary['Salary'])
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=15)
plt.title('Department wise salary')
#plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
for i, val in enumerate(Total_Salary['Salary']):
    plt.text(Total_Salary['Department'][i], val, str(val))

plt.show()
```

```python
data
```

```python
data.columns
```

```python
emp_detail= data.groupby('Department').agg(
    total_salary= ('Salary','sum'),
    total_absent = ('Absences','sum'),
    total_employees = ('EmpID','count'),
    total_satis = ('EmpSatisfaction','sum')
    
).reset_index()
emp_detail
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))
plt.suptitle("Department vs Other Features")
plt.subplots_adjust(hspace=0.5)

ax1 = axes[0,0]
ax1.bar(emp_detail['Department'], emp_detail['total_salary'])
ax1.set_title("Department vs Salary")
ax1.set_xlabel('Department')
ax1.set_ylabel('Total Salary')
ax1.tick_params(axis ='x',rotation=15)
for i, val in enumerate(emp_detail['total_salary']):
    ax1.text(emp_detail['Department'][i], val+3, str(val), ha='center')
    
ax2 = axes[0,1]
ax2.bar(emp_detail['Department'], emp_detail['total_absent'])
ax2.set_title("Department vs Absent")
ax2.set_xlabel('Department')
ax2.set_ylabel('Total_absents')
ax2.tick_params(axis ='x',rotation=15)
for i, val in enumerate(emp_detail['total_absent']):
    ax2.text(emp_detail['Department'][i], val+3, str(val), ha='center')
ax3 = axes[1,0]
ax3.bar(emp_detail['Department'], emp_detail['total_employees'])
ax3.set_title("Department vs Salary")
ax3.set_xlabel('Department')
ax3.set_ylabel('Total_employees')
ax3.tick_params(axis ='x',rotation=15)
for i, val in enumerate(emp_detail['total_employees']):
    ax3.text(emp_detail['Department'][i], val+3, str(val), ha='center')
ax4 = axes[1,1]
ax4.bar(emp_detail['Department'], emp_detail['total_satis'])
ax4.set_title("Department vs Satisfaction")
ax4.set_xlabel('Department')
ax4.set_ylabel('Satisfaction')
ax4.tick_params(axis ='x',rotation=15)
for i, val in enumerate(emp_detail['total_salary']):
    ax3.text(emp_detail['Department'][i], val+3, str(val), ha='center')

```
