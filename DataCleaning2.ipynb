```python
item_weight_mean = data['Item_Weight'].mean()
print(f"{item_weight_mean:.2f}")
```

```python
data['Item_Weight'] =data['Item_Weight'].fillna(item_weight_mean)
data['Item_Weight']
```

```python
outlet_size_mode = data['Outlet_Size'].mode()[0]
```

```python
outlet_size_mode
```

```python
data['Outlet_Size']=data['Outlet_Size'].fillna(outlet_size_mode)
```

```python
data['Outlet_Size']
```

```python
data.isnull().sum()
```

```python
data['Outlet_Establishment_Year'] = pd.to_datetime(data['Outlet_Establishment_Year'])
```

```python
data.info()
```

```python
#Removing white spaces in Item_Identifier coloumn
data['Item_Identifier']=data['Item_Identifier'].str.strip()
```

```python
data['Item_Identifier']
```

```python
#Mapping for Outlet_Type coloumn
data['Outlet_Type'].unique()
```

```python
outlet = {
    "Supermarket Type1":"Supermarket",
    "Supermarket Type2":"Supermarket",
    "Supermarket Type3":"Supermarket",
    "Grocery Store":"Grocery Store"
}
```

```python
data['Outlet_Type'] =data['Outlet_Type'].map(outlet)
```

```python
data['Outlet_Type']
```

```python
#Typecasting coloumn Item_Outlet_Sales
data['Item_Outlet_Sales']=data['Item_Outlet_Sales'].astype(int)
```
