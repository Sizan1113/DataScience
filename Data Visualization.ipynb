```python
import matplotlib.pyplot as plt
```

Line Plot

```python
fruits = ['Apple','Banana','orange','Coconut']
sales = [100,500,350,200]
revenue =[150,100,500,300]
 #X->independent variable
#y ->dependent variable
```

```python
plt.figure(figsize = (6,4))
plt.plot(fruits,sales,color='red' ,marker='o', label = 'Sales' )
plt.plot(fruits,revenue,color = 'green',marker = '*',label = 'Revenue',linestyle='--')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits and their sakes')
plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
plt.grid(True,linestyle='--', color = 'Blue', alpha=0.9)
plt.savefig('Fruits_sales.png')
plt.show()
```

Bar Chart

```python

plt.bar(fruits,sales,edgecolor= 'black' ,color=['orange','green','blue','red'])
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.title('Fruits and their sales')
#plt.legend(loc='upper right', bbox_to_anchor=(1.4,1))
plt.grid(True,linestyle='--', color = 'Blue', alpha=0.3)
plt.savefig('Fruits_sales.png')
plt.show()
```
